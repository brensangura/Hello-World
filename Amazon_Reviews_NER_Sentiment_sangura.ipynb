```python
import spacy

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Define sentiment rules
positive_words = ["good", "great", "excellent", "love", "awesome", "fantastic", "recommend", "best"]
negative_words = ["bad", "poor", "terrible", "awful", "hate", "worst", "waste", "disappointing"]

def analyze_review(review):
    doc = nlp(review)

    # Task 1: Extract PRODUCT and ORG entities
    entities = []
    for ent in doc.ents:
        if ent.label_ in ["PRODUCT", "ORG"]:  # PRODUCT=product names, ORG=brands
            entities.append((ent.text, ent.label_))

    # Task 2: Rule-based sentiment analysis
    positive_count = 0
    negative_count = 0

    for token in doc:
        word = token.text.lower()
        if word in positive_words:
            positive_count += 1
        elif word in negative_words:
            negative_count += 1

    # Determine overall sentiment
    sentiment = "positive" if positive_count > negative_count else "negative"

    return entities, sentiment

# Example Amazon review
review = """
I bought an iPhone 13 from Apple. The battery life is great, but the screen is poor.
Overall, it's a good product though.
"""

# Analyze the review
entities, sentiment = analyze_review(review)

# Print results
print("--- Extracted Entities ---")
for entity, label in entities:
    print(f"- {entity} ({label})")

print("\n--- Sentiment Analysis ---")
print(f"Sentiment: {sentiment.capitalize()}")
```
